```
# File location and type
file_location = "/FileStore/tables/p1p2Pub_doi.csv"
file_type = "csv"

# CSV options
infer_schema = "false"
first_row_is_header = "false"
delimiter = ","

# The applied options are for CSV files. For other file types, these will be ignored.
df = spark.read.format(file_type) \
  .option("inferSchema", infer_schema) \
  .option("header", first_row_is_header) \
  .option("sep", delimiter) \
  .load(file_location)

display(df)
```

```
df.createOrReplaceTempView('df')
```

```
Papers = MAG.getDataframe('Papers')
PaperDoi = Papers.select(Papers.PaperId, Papers.Doi)
PaperDoi.show(10)
PaperDoi.createOrReplaceTempView('PaperDoi')
```

```
%sql

SELECT *
FROM (SELECT _c6 AS unique_identifier
      FROM df) AS DF
LEFT JOIN PaperDoi 
ON DF.unique_identifier = PaperDoi.Doi
-- LIMIT 10
```

```
Authors = MAG.getDataframe('Authors')
Authors = Authors.select(Authors.AuthorId, Authors.DisplayName, Authors.LastKnownAffiliationId, Authors.PaperCount)
Authors.show(3)
Authors.createOrReplaceTempView('Authors')
```

```
# Get (author, paper) pairs
PaperAuthorAffiliations = MAG.getDataframe('PaperAuthorAffiliations')
AuthorPaper = PaperAuthorAffiliations.select(PaperAuthorAffiliations.AuthorId, PaperAuthorAffiliations.PaperId).distinct()
AuthorPaper.show(3)
AuthorPaper.createOrReplaceTempView('AuthorPaper')
```

```
# File location and type
file_location = "/FileStore/tables/Full_mag_phase1_2_doi-31729.csv"
file_type = "csv"

# CSV options
infer_schema = "false"
first_row_is_header = "false"
delimiter = ","

# The applied options are for CSV files. For other file types, these will be ignored.
doi = spark.read.format(file_type) \
  .option("inferSchema", infer_schema) \
  .option("header", first_row_is_header) \
  .option("sep", delimiter) \
  .load(file_location)

display(doi)
```

```
doi.createOrReplaceTempView('doi')
```

```
%sql 
select ap.PaperId, a.*
from AuthorPaper as ap
left join Authors as a
on a.AuthorId = ap.AuthorId
where ap.PaperId in (
  select _c1
  from doi
)
```
