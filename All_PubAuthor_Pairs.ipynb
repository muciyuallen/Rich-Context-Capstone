```python
import networkx as nx
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import time

%matplotlib inline
```

```python
pub_author = pd.read_csv('data/Full_mag&publications_Authors.csv', dtype=object )
p1p2_author = pd.read_csv('data/Full_mag&phase1+2_Authors.csv', dtype=object )
```

```python
pub_author.head()
```

```python
p1p2_author.head()
```

```python
all_author = pd.concat([p1p2_author, pub_author])
all_author.head(10)
```

```python
all_doi = pd.read_csv('data/allPub_doi.csv', dtype=object )
all_doi.head()
```

```python
all_paperID = pd.read_csv('data/alldata_doi_paperID.csv', dtype=object )
```

```python
all_paperID.head(10)
```

```python
all_id_doi = pd.merge(all_doi, all_paperID, left_on='unique_identifier', right_on='Doi', how='inner')[['publication_id','PaperId','Doi']]
all_id_doi.head()
```

```python
all_PubAuthor_pairs = pd.merge(all_id_doi, all_author, left_on='PaperId', right_on='PaperId', how='inner')[['publication_id','AuthorId','Doi']]
all_PubAuthor_pairs.head(10)
```

```python
all_PubAuthor_pairs.to_csv('all_PubAuthor_pairs.csv')
```
